### Posible montaje para la toma de datos del punto (g)
Montaje_Arduino.jpg

```cpp
// Arduino code
//
double VC = 0.0;

void setup()
{
    Serial.begin(9600);
    pinMode(4, OUTPUT);
    pinMode(A0, INPUT);
}

void loop()
{
    digitalWrite(4, HIGH);
    while (abs(VC - 5)>1e-2) {
        VC = analogRead(A0) * 5.0 / 1023.0;
        Serial.println(VC);
    }
    delay(10); // Delay a little bit to improve simulation performance
    digitalWrite(4, LOW);
    while (VC > 1e-2) {
        VC = analogRead(A0) * 5.0 / 1023.0;
        Serial.println(VC);
    }
    delay(10);
}
```

## Caso 1: archivo simple de una columna
En este ejemplo se leerá un archivo de texto que se encuentra en la memoria de la computadora, y que contiene una sola columna. La lectura del archivo se hará usando la librería ``pandas``.

```python
### Importar las librerías ##
# El primer paso es importar las librerías que se necesitan para leer el archivo y graficar con los datos un histograma,
# determinar la media y la desviación estándar
import numpy as np # para determinar la media y la desviación estándar
import pylab as plt # para graficar el histrograma
import pandas as pd # para leer el archivo de texto con los datos
```

```python
### Leer el archivo de datos ##

#--- Este código permite leer la memoria del computador desde Colaboratory ---
from google.colab import files
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
#--- +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++ ---

# se utiliza la libreria pandas, pero existen otras formas de leer este tipo de archivos. 
# Para la lectura necesitamos el nombre del archivo, y precisar que la primera fila No corresponde
# a los nombres de las columnas

df = pd.read_csv("Datos_sonido.txt",header=None) # Se llama el arvhivo y los datos se guardan en un dataframe
print(df)
```

Note que el dataframe automáticamente le asignó un numero a cada columna y fila para identificarlas.

```python
### Se crea el arreglo de datos para poder hacer operaciones algebraicas
#señal = df[0].as_matrix() # Forma de versiones antiguas de pandas
señal = df[0].to_numpy() # Forma de versiones modernas de pandas
print(señal.dtype)
print(señal)
```

La operación anterior convirtió los datos en un arreglo de datos de tipo int64, enteros de 64 bit.

```python
### Grafica del histograma ##
plt.figure()
n, bins, patches = plt.hist(señal,bins=int(np.sqrt(len(señal))),edgecolor= 'k',color='red', alpha=0.5)
plt.ylabel('Frecuencia')
plt.xlabel('Señal')
plt.show()
print(n,bins,patches)
```

La información que nos entrega la función ``hist()`` es $\text{n}$, que corresponde a la frecuencia de cada "bin" o intervalo definido, $\text{bins}$, que nos indica el rango de cada intervalo, y $\text{patches}$, que nos dice el número de intervalos generados.

```python
### Determinación de la media y la desviación estándar ##
media = np.mean(señal)
desv = np.std(señal)
print('Media de la distribución = %.1f'%media)
print('Desviación estándar de la distribución = %.1f'%desv)
```

## Asignación automatica de tipos de variables

En Python 2 no es necesario declarar explícitamente el tipo de las variables. El intérprete trata de inferir el tipo de las variables según se usan. Igualmente, las operaciones tienen semántica distinta para distintos tipos de datos.  Pero en Python 3 las operaciones tipo fracción son float independiente del tipo de variable original.

```
a = 10
b = 3
c = 3.
d = "hola"

print("var \t value \t type")
print("...............................")
print("a \t",a, "\t", type(a))
print("b \t",b, "\t", type(b))
print("c \t",c, "\t", type(c))
print("d \t",d, "\t", type(d))
print("...............................\n")

print("la división siempre da un float...")
print("a/b =",(a/b))
print("a/c =", a/c)
```

## Listas

Las listas pueden tener cualquier tipo de elemento

```
a = [1, 2, 3, "hola", [10, "nunca", 90], -32]
print(a)
print("tamaño de la lista ->",len(a)) #tamaño de la lista
```

Para recorrer los elementos de una lista con un **For**:

```
print("indx \t element")
for i in range(len(a)):
    print(i, "\t", a[i])
```

Note que el índice del primer elemento de la lista es 0 y no 1.
Debido a que las listas de python son una clase iterable, también es posible recorrer sus elementos como sigue:

```
for i in a:
    print(i)
```

También es posible extrear porciones de datos de una lista (sublistas). Si se quisiera obtener solo los datos de la lista $a$ desde el índice $i$ hasta el $i+20$ se escribiría $a[i:i+20]$. O si solo quisiera excluir de su lista los ultimos cinco valores podría hacerlo escribiendo $a[:-5]$.

```
#todos a partir del elemento en la tercera posición
print(a[2:])
 #desde la antepenúltima hasta la última
print(a[-3:])
#Otra forma de escribir la de arriba
print(a[-(len(a)-3):]) 
#elemento en la posición 5
print(a[4])
#los elementos de la lista dando pasos de 2.
print(a[::2])
```

```
print(a[4][1]) #segundo elemento del elemento en la posición 5
print(a[4][1:]) #a partir del segundo elemento del elemento en la posición 5
```

Tenga en cuanta que los objetos tipo "string" también son objetos iterables, por tanto se pueden extraer sus caracteres al igual que los elementos de una lista

```
print(a[3])
# las dos ultimas letras de la palabra hola, que es el cuarto elemento de a.
print(a[3][2:])
```

```
print(a[5])
print(a[5][1:])
# El error se produce porque el número no se puede separar.
```

Con la notación de índices accedemos a columas o filas enteras, rangos de columnas o filas, elementos individuales o porciones de una matriz o un vector.

## Operaciones vectorizadas

```
# los array de numpy admiten varias operaciones
v = np.array([10,12,13,15,20])
print(v)
print(v+1) # sumar un escalar a todas las entradas
print(v*2) # multiplicar tosas las entradaas por un escalar
print(v.dot(v)) # producto interno entre dos "vectores"
```

```
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[6,5,4],[3,2,1]])
c = np.array([[1,2],[4,5]])
print(a)
print('')
print(b)
print("--")
print(a.T.dot(a)) #Esta es una forma de realizar producto matricial, porque numpy por defecto realiza producto punto a punto
print('')
print(a*b)
```

```
a = np.array([[1,2,3],[4,5,6]])
b = np.array([[6,5,4],[3,2,1]])

print(a)
print('')
print(b)
print("--")

print("a+b\n",a+b)
print("a**2\n", a**2)
print("a*b\n",a*b) # no es producto matricial !
print("a x b'\n",a.dot(b.T))
print("a' x b\n",a.T.dot(b))
```

```
# es posible realizar operaciones booleanas y usarlas para acceder a ciertos 
# elementos de un array
a = np.array([1,8,4,10,-4,5])
print(a)
print(a>4) #esta línea retorna una respuesta booleana a la pregunta

a[a>4] #esta línea retorna una respuesta numérica

i = np.array([False, True, False, True, False, True])
print(i)
```

Las operaciones vectorizadas también funcionan con expresiones _booleanas_. Fíjate cómo se indexa un vector con una expresión booleana para seleccionar un conjunto de elementos.

```
a = np.array([1,8,4,10,-4,5])
print("posiciones en a >4:", a>4)
print("elementos de a >4:",a[a>4])
```

## Funciones y estructuras

Python es un lenguaje **indentado**, no usa corchetes para delimitar el alcance de las estructuras de programación sino que se fija en los **cambios de indentación**. 

No se declara el tipo de los argumentos de las funciones. La semática de la implementación ha de estar preparada para funcionar con los tipos de datos que quieres.

```
def funcion_1(a,b):
    # todo lo que este identado hace parte de la definición de la función
    r = a**2
    return r+b
    
def greatest(a,b):
    if a>b:
        return a
    else:
        return b

m1 = np.array([[3,4],[1,1]])
m2 = np.array([[5,6],[0,0]])
print(funcion_1 (10.4,2))
print('')
print(funcion_1 (10.4, np.array([2,4])))
print('')
print(funcion_1 (m1,m2))
print('')
print(greatest(10,2))
```

Podemos definir valores por defecto para los argumentos de las funciones y llamarlas usando explícitamente el nombre de los argumentos. Además, las funciones pueden devolver varios valores.

```
def f_power(x, p=2):
    return x**p

print(f_power(x=3))
print(f_power(p=4, x=3))
print(f_power(4,3)) #note que si no se define la variable se deben ingresar en el orden correcto
print(f_power(3,4))
```

```
def f_power(x, p=2):
    return x**p, x*p

r = f_power(p=4, x=3)
print('r',r[1])

r1, r2 = f_power(p=4, x=3)
print('r1',r1)
print('r2',r2)


a,b = 10, np.array([10,4,-3])
print(a)
print(b)
```

```
def f_power(x, p=2):
    return x**p

def f_powers(x, p1=2, p2=3):
    return x**p1, x**p2

print(f_power(4))
print(f_power(4,3))
print(f_powers(4, p1=3))
xp1, xp2 = f_powers(4, p2=4, p1=3)
print("power1",xp1, "power2", xp2)
```

## Archivos de texto txt

Para realizar un ejemplo de lectura de archivos, primero generamos uno con los datos de una distribución normal.


```
x = np.random.normal(loc=5, scale=1.2, size=(100))
```

```
#note que no hay un archivo con el nombre "datos.txt"
!ls
```

```
# para escribir un archivo...

file = open("datos.txt", 'w')

for i in x:
  file.write("%.4f\n"%i)

file.close()

#Ahora si hay un archivo con el nombre "datos.txt"
!ls
```

para leer el archivo de datos se puede usar la función `loadtxt` de `numpy`, pero existen otras librerías especializadas en análisis de datos como `pandas`.

```
datos = np.loadtxt("datos.txt")
datos # ya acá se tienen los datos disponibles para los analsis
```

```
### Grafica del histograma ##
plt.figure()
n, bins, patches = plt.hist(datos,bins=int(np.sqrt(len(datos))),edgecolor= 'k',color='red', alpha=0.5)
plt.ylabel('Frecuencia')
plt.xlabel('x')
plt.show()
print(n,bins,patches)
```

La información que nos entrega la función ``hist()`` es $\text{n}$, que corresponde a la frecuencia de cada "bin" o intervalo definido, $\text{bins}$, que nos indica el rango de cada intervalo, y $\text{patches}$, que nos dice el número de intervalos generados.

```
### corroboramos la media y la desviación estándar ##
media = np.mean(datos)
desv = np.std(datos)
print('Media de la distribución = %.1f'%media)
print('Desviación estándar de la distribución = %.1f'%desv)
```

## Imágenes

```
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
from scipy.signal import convolve2d
%matplotlib inline

url1 = "https://static.iris.net.co/semana/upload/images/2017/10/26/544975_1.jpg"
img = io.imread(url1)
#img = io.imread("imgs/sample_img.jpg") #esta línea es si la imagen está en una carpeta. No funciona directamente para Colaboratory
print("dimensiones", img.shape, "max", np.max(img), "min", np.min(img))
plt.scatter(320,60, marker="x", s=200, linewidth=5, c="b")
plt.scatter(150,230, marker="x", s=200, linewidth=5, c="r")
print("pixel at blue marker ", img[60,320,:])
print("pixel at red marker", img[230,150,:])
plt.grid() # remove gridlines
plt.imshow(img)
type(img)
```

```
plt.imshow(img[:,:,0], cmap = plt.cm.Greys_r)
```

La manipulación de imágenes se _reduce_ a realizar cálculos sobre los valores de luminosidad de cada pixel. Por ejemplo, obtenemos una versión en escala de grises promediando los valores RGB de cada pixel. Esto se hace de manera natural con la función `np.mean` y el argumento `axis` adecuado.

```
gimg = np.mean(img, axis=2)
print("dimensiones", gimg.shape, "max", np.max(gimg), "min", np.min(gimg))
plt.scatter(320,60, marker="x", s=200, linewidth=5, c="b")
plt.scatter(150,230, marker="x", s=200, linewidth=5, c="g")
print("pixel at blue marker ", gimg[60,320])
print("pixel at green marker", gimg[230,150])
plt.grid() # remove gridlines
plt.imshow(gimg, cmap = plt.cm.Greys_r, vmin=0, vmax=255)
```

También podemos acceder a porciones (**parches**) de la imagen usando la notación natural de matrices de Python. Si además, reducimos la luminosidad de cada pixel a la mitad lo que hacemos es oscurecer la imagen.

```
parche = gimg[50:100, 300:350]
parche_oscuro = parche/2

fig = plt.figure(figsize=(3,3))
fig.add_subplot(121); plt.grid(); 
plt.imshow(parche, cmap=plt.cm.Greys_r)
fig.add_subplot(122); plt.grid(); 
plt.imshow(parche_oscuro, cmap=plt.cm.Greys_r)
```

```
np.mean(parche), np.mean(parche_oscuro)
```
